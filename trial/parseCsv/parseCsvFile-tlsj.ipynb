```python
# ========== 根据上一步得到的代码列表，遍历所有股票，将这些股票合并到一张表格 all_stock 中
all_stock = pd.DataFrame()

# 从csv文件中读取该股票数据 
# 注意：这里请填写数据文件在您电脑中的路径
stock_data = pd.read_csv(input_stock_data_csv_path,encoding='gbk')
# print stock_data.columns
# 选取需要的字段，去除其他不需要的字段
#"timestamp","ticker","exchangeCD","shortNM","lastPrice"
stock_data = stock_data[[ u'ticker',u'exchangeCD', u'shortNM',u'lastPrice']]
stock_data.columns = ['code','market','shortname','price']
```

```python
def getStockCode(x):
    if len(str(x)) == 6 and x[0:3].isnumeric() and \
        (int(x[0:3]) == 600 or int(x[0:3]) == 601 or (x[0:2]) == "00" or int(x[0:2]) == 30):
        return x
    else:
        return np.nan
```

```python
def getMarketCode(x):
    if x == "XSHE" or x == "XSHG":
        return x
    else:
        return np.nan
```

```python
# 去掉 stock 代表市场的前两个字符
trim_stock_code = stock_data['code'].map(getStockCode)
trim_market_code = stock_data['market'].map(getMarketCode)
stock_data['code2'] = trim_stock_code
stock_data['market2'] = trim_market_code
stock_data = stock_data.dropna()
stock_data
```

```python
all_record = []
all_num = 0
left_num = 0
for id,row in stock_data.iterrows():
    all_num += 1
    if  row['market'] != 'XSHG' and row['market'] != 'XSHE':
        print "market error:", row['code'],row['shortname']
        row['code2'] = np.nan
        continue
        
    if int(row['code2'][0]) == 6 and row['market'] != 'XSHG':
        print "market 6 XSHG error:", row['code'],row['shortname']
        row['code2'] = np.nan
        continue

    if (int(row['code2'][0]) == 0 or  int(row['code2'][0]) == 3 )and row['market'] != 'XSHE':
        print "market 0 3 XSHE error:", row['code'],row['shortname']
        row['code2'] = np.nan
        continue

    if row['shortname'].encode('UTF-8','ignore').find('指数') != -1:
        print "指数 error:", row['code'],row['shortname']
        row['code2'] = np.nan
        continue

    if row['shortname'].encode('UTF-8','ignore').find('上证') != -1:
        print "上证 error:", row['code'],row['shortname']
        row['code2'] = np.nan
        continue 
    if row['shortname'].encode('UTF-8','ignore').find('深证') != -1:
        print "深证 error:", row['code'],row['shortname']
        row['code2'] = np.nan
        continue
    if row['shortname'].encode('UTF-8','ignore').find('成指') != -1:
        print "成指 error:", row['code'],row['shortname']
        row['code2'] = np.nan
        continue
    if row['shortname'].encode('UTF-8','ignore').find('综指') != -1:
        print "综指 error:", row['code'],row['shortname']
        row['code2'] = np.nan
        continue        
        
    if int(row['price']) > 1000:
        print " > 1000 error:", row['code'],row['shortname']
        row['code2'] = np.nan
        continue    
        
    if float(row['price']) < 0.1:
        print " < 0.1 error:", row['code'],row['shortname']
        row['code2'] = np.nan
        continue            
        
    #print row
    left_num += 1
    all_record.append(row)
        
        
```
