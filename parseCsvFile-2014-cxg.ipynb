```python
col_num = len(stock_data.columns)
```

```python
#first function ,spend time lots
all_stock = pd.DataFrame()
i=0
while i <= col_num : 
    date=stock_data.columns[i]
    code=stock_data.columns[i + 1]
    code_cn=stock_data.columns[i + 2]
    print date,code,code_cn
    #
    for idx, row in stock_data.iterrows():
        # 选取需要的字段，去除其他不需要的字段
        if len(str(row[code])) > 4 :
            stock_record = pd.Series([date,row[code],row[code_cn]])
            all_stock = all_stock.append(stock_record, ignore_index=True)
    #    
    i+=5
```

```python
#second function,spends times low
all_stock = pd.DataFrame()
for idx, row in stock_data.iterrows():
    i=0
    while i <= col_num : 
        date=stock_data.columns[i]
        code=stock_data.columns[i + 1]
        code_cn=stock_data.columns[i + 2]
        #print date,code,code_cn
        # 选取需要的字段，去除其他不需要的字段
        if len(str(row[code])) > 4 :
            stock_record = pd.Series([date,row[code],row[code_cn]])
            all_stock = all_stock.append(stock_record, ignore_index=True)
        i+=5
```

```python
# 添加列名称
all_stock.columns=['date','code','code_cn']
```
