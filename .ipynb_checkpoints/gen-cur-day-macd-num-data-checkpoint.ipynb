```python
from __future__ import division
from numpy.random import randn
from pandas import Series
import numpy as np
import pandas as pd
import os
import time;
import datetime  
np.set_printoptions(precision=4)
import sys; sys.path.append('book_scripts')
%pwd
```

```python
# input CSV数据文件
input_old_stock_data_path = 'E:/project/pychram/traderesp/base/input-csv/20151101-20151218-history-macd/20151101-20151218-history-macd.csv'
# input everyday-stock path
input_ervery_day_stock_path = 'E:/project/pychram/traderesp/base/input-csv/everyday-stock-history'
```

```python
cur_day = datetime.datetime.now() 
cur_day_str = cur_day.strftime('%Y-%m-%d');  
cur_day_str 
```

```python
all_stock = pd.DataFrame()
count = 0;
load_count = 0;
while count < 50:
    count = count + 1
    prev_day = cur_day + datetime.timedelta(days = -count)
    prev_day_str = prev_day.strftime('%Y-%m-%d'); 
    prev_day_path_str = input_ervery_day_stock_path + "/overview-push-" + prev_day_str
    #print prev_day_path_str
    if os.path.isdir(prev_day_path_str):
        load_count = load_count + 1
        file_name = prev_day_path_str + "/stock overview.csv"
        print file_name
        stock_data = pd.read_csv(file_name,parse_dates=[2],encoding='gbk')
        # print stock_data.columns
        # 选取需要的字段，去除其他不需要的字段
        # 股票代码,股票名称,交易日期,新浪行业,新浪概念,新浪地域,开盘价,最高价,最低价,收盘价,后复权价,前复权价,涨跌幅,成交量,成交额,换手率,流通市值,总市值,是否涨停,是否跌停,市盈率TTM,市销率TTM,市现率TTM,市净率,MA_5,MA_10,MA_20,MA_30,MA_60,MA金叉死叉,MACD_DIF,MACD_DEA,MACD_MACD,MACD_金叉死叉,KDJ_K,KDJ_D,KDJ_J,KDJ_金叉死叉,布林线中轨,布林线上轨,布林线下轨,psy,psyma,rsi1,rsi2,rsi3
        stock_data = stock_data[[ u'交易日期',u'股票代码', u'收盘价',u'后复权价','MACD_DIF','MACD_DEA','MACD_MACD']]
        stock_data.columns = ['date','code','close','back_close','macd_dif','macd_dea','macd_macd']
        # 去掉 stock 代表市场的前两个字符
        trim_stock_code = stock_data['code'].map(lambda x: x[2:])
        stock_data['code'] = trim_stock_code
        #
        # 将该股票的合并到output中
        all_stock = all_stock.append(stock_data, ignore_index=True)
if load_count < 30:   
    old_file_name = 'E:/project/pychram/traderesp/base/input-csv/20151101-20151218-history-macd/20151101-20151218-history-macd.csv'
    stock_data = pd.read_csv(old_file_name,parse_dates=[2],encoding='gbk')
    # 将该股票的合并到output中
    all_stock = all_stock.append(stock_data, ignore_index=True)   
```
